TO CHECKK WHEATHER WE HAVE CLEANED OUR DATA PROPERLU OR NOT

```python
#Now we will plot dispot?histplot to check propely
# here displot only work with numeric data

df4=df3.select_dtypes(include="number").columns.to_list()
print(df4)
```

```python
# original data
sns.displot(df["MSSubClass"],kde=True)
```

```python
# cleaned data displot
sns.displot(df3["MSSubClass"],kde=True)
```

```python
#lets overlap it
sns.histplot(df["MSSubClass"],kde=True,label="original")
sns.histplot(df3["MSSubClass"],kde=True,label="cleaned")
```

```python
#One by one all numeric plot
plt.figure(figsize=(25,25))
for i,var in enumerate(df4):
  plt.subplot(10,4,i+1)
  sns.histplot(df[var],bins=20,kde=True, label="Original")
  sns.histplot(df3[var],bins=20,kde=True,label="Cleaned")
  plt.legend()
```

Now categorical data

```python
df5=df3.select_dtypes(include="object").columns.to_list()
print(df5)
len(df5)
```

```python
#lets check of single variable
df_check=pd.concat([df["Street"].value_counts()/df.shape[0]*100,
           df3["Street"].value_counts()/df3.shape[0]*100],axis=1,keys=["Original","Cleaned"])
df_check["% Change"] = df_check["Cleaned"] - df_check["Original"]
print(df_check)

```

```python
#lets check for all categorical data
def check(df5):
  for i in df5:
    df_check=pd.concat([df[i].value_counts()/df.shape[0]*100,
                        df3[i].value_counts()/df3.shape[0]*100],axis=1,
                       keys=["Original","Cleaned"])
    df_check["% change"]=df_check["Cleaned"]-df_check["Original"]
    print(df_check)

check(df5)
```
